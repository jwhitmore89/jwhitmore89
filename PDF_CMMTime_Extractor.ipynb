```python
import os
import numpy as np
import PyPDF2
from PyPDF2 import PdfFileReader
import csv
import time

dwg_cmm_list = [["Drawing Number","Date","Time","Plan Name","Job","Operation","CMM No.","CMM Time","QA Process","Run"]]
dwg_time_list = [["Drawing Number","Execution Time"]]
```

```python
'''
This program is designed to extract useful data from PDF files and store the data in a CSV file.

The program takes 'folder' as an input and will iterate through each file in the folder and collect relavant information.

Information currently extracted includes 'Drawing Number', 'Plan Name', 'Date', 'Time', 'Opperation', and 'QA Process'. These 
variables can commented out or restored as necessary.

Program has funtionality to report Execution Time, current page number, confirmation of file closure, and confirmation of 
data extraction.

Program may not function propperly for PDF's created circa 2018 and before due to formatting issues.
'''








## Record Start time
start_time = time.time()

## Open file and begin file stream (file is opened to stream rather than imported)
pdf_file = open("easy_name8.pdf","rb")
read_pdf = PdfFileReader(pdf_file)

## Initialize desired variables
drawing_no = None
cmm_time = None
plan_name = None
date = None
time = None

page_num = 0

## Begin while loop to search each page for relavent variables
## Page iterates at end of loop if any variable(s) is still defined as type 'None'

while drawing_no == None or cmm_time == None:

    page = read_pdf.getPage(page_num)
    text = page.extractText()
    text_list = text.split("\n")

    for line in text_list:
        if "Drawing No." in line:
            drawing_line = text_list.index(line)
            drawing_no = text_list[drawing_line + 1]

        if "CMM Time" in line:
            cmm_line = text_list.index(line)
            cmm_time = text_list[cmm_line + 1]       
    print(page_num)
    page_num += 1
    
print(text) 
    
    
## Optional, Debugging: Print currently stored Drawing number or CMM time
'''
print("Drawing No. is ", drawing_no, "\n", "CMM Time is ", cmm_time) 
'''
## Stores extracted values in  a list and appends them to a master list for the current run

dwg_cmm = [drawing_no,cmm_time]
dwg_cmm_list.append(dwg_cmm)

## Stores execution time and appends to list with respective drawing number
'''
exec_time = "--- %s seconds ---" % (time.time() - start_time)
dwg_time_list.append([drawing_no,exec_time])
'''
## Close opened file

pdf_file.close()
print(pdf_file.closed)

## Optional: Print execution time and 'finished'
'''
print("--- %s seconds ---" % (time.time() - start_time))
print("Finished")
'''
```
